```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# initialize the grid
A = np.array([[0.5,0],[0,2]])
# x_start = np.array([-2,-2]).astype(float)
x_start = np.array([0.1,0.1]).astype(float)
# x_start = np.array([-2,-1]).astype(float)
x_goal = np.array([1.194,-1.675]).astype(float) #numerical approximation of minimum
x1 = np.linspace(-4, 4, 200)
x2 = np.linspace(-4, 4, 200)
X1,X2 = np.meshgrid(x1, x2)
X = np.vstack((X1.reshape((1,-1)),X2.reshape((1,-1))))
X.shape
```

### Loss Function
$L=\left(1-x_0^{2}+x_0x_1 ^{3}\right) e^{-x_0^{2} / 2-x_1^{2} / 2}+\frac{1}{100}(x_0-2)^{2}+\frac{1}{100}(x_1+2)^{2}+0.72$

### Gradient
- $\frac{\partial{L}}{\partial{x_0}} = e^{-x_0^{2} / 2-x_1^{2} / 2}\left(x_0^{3}-x_0^{2} x_1^{3}-3 x_0+x_0^{3}\right)+\frac{x_0-2}{50}$

- $\frac{\partial{L}}{\partial{x_1}} = y e^{-x^{2} / 2-y^{2} / 2}\left(x^{2}-x y^{3}+3 x y-1\right)+\frac{y+2}{50}$

https://www.wolframalpha.com/input/?i=d%2Fdx+(1-x**2%2Bx*y**3)*e%5E(-x**2%2F2-y**2%2F2)+%2B+(x-2)**2%2F100+%2B+(y%2B2)**2%2F100+%2B+0.72

```python
# # define loss function and compute loss for the grid
# def loss_fcn(W, x):
#     return 0.5*x[0]*x[1]*np.exp(-x[0]**2-x[1]**2)
# # define the gradient calculation
# def gradient(W, x):
#     return np.array([x[1]*(0.5-x[0]**2)*np.exp(-x[0]**2-x[1]**2),
#                      x[0]*(0.5-x[1]**2)*np.exp(-x[0]**2-x[1]**2)])
def loss_fcn(x):
    return (1-x[0]**2+x[0]*x[1]**3)*np.exp(-x[0]**2/2-x[1]**2/2) + (x[0]-2)**2/100 + (x[1]+2)**2/100 + 0.72
def gradient(x):
    temp = np.exp(-x[0]**2/2-x[1]**2/2)
    return np.array([(-3*x[0]+x[1]**3+x[0]**3-x[0]**2*x[1]**3)*temp + (x[0]-2)/50,
                    (3*x[0]*x[1]**2-x[1]+x[0]**2*x[1]-x[0]*x[1]**4)*temp + (x[1]+2)/50])
Y = np.array([])
for x in X.T:
    Y = np.append(Y, loss_fcn(x))
y_start = loss_fcn(x_start)
y_goal = loss_fcn(x_goal)

delta = 0.001
temp1 = x_start.copy()
temp1[0]+=delta
temp2 = x_start.copy()
temp2[1]+=delta
print("Numerical approximation of gradient: ", [(loss_fcn(temp1)-loss_fcn(x_start))/delta, (loss_fcn(temp2)-loss_fcn(x_start))/delta])
print("Derived Gradient: ",gradient(x_start))
# # Numerical minimum
# X.T[Y.argmin()]
# Y.min()
```

```python
# 3d plot of the cost functions
from mpl_toolkits.mplot3d import axes3d
def plot_loss_3d():
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    ax.contour3D(X1, X2, Y.reshape(X1.shape),100)
    start_plt = ax.scatter(x_start[0],x_start[1],y_start,s=50,c='r', marker = "o")
    goal_plt = ax.scatter(x_goal[0],x_goal[1],y_goal,s=50,c='r', marker = "x")
    ax.set_xlabel('$\mathregular{x_1}$')
    ax.set_ylabel('$\mathregular{x_2}$')
    ax.set_zlabel('y');
    ax.view_init(30, 40)
    ax.axis('equal')
    plt.legend([start_plt, goal_plt], ['start', 'goal'])
    ax.axis([-5,5,-5,5])
    return ax
plot_loss_3d()
plt.savefig("loss_fcn_plot_3d.png")
# 2D contour plot of the cost function
def plot_loss_2d():
    plt.figure(figsize=(10,10))
    levels = np.linspace(0,1.3, 8)
    cp = plt.contour(X1,X2,Y.reshape(X1.shape), colors='k', linestyles='dashed', levels=levels)
    plt.clabel(cp, inline=1, fontsize=10)     
    cp = plt.contourf(X1, X2, Y.reshape(X1.shape))
    start_plt = plt.scatter(x_start[0],x_start[1], s=50, c='r', marker = "o")
    goal_plt = plt.scatter(x_goal[0],x_goal[1], s=50, c='r', marker = "x")
    plt.xlabel('$\mathregular{x_1}$')
    plt.ylabel('$\mathregular{x_2}$')
    plt.axis('equal')
    plt.legend([start_plt, goal_plt], ['start', 'goal'])
plot_loss_2d()
plt.savefig("loss_fcn_plot_2d.png")
```

```python
# Gradient Descent
lr = 1

# stopping criteria
max_iter = 100
iter_count = 0
grad_thres = 1e-5

#initialization
x = x_start.copy()
x_hist_gd1 = x.reshape(2,1).copy()
y_hist_gd1 = np.array([y_start])
grad_mag = float('inf')
while iter_count<max_iter and grad_mag>grad_thres:
    iter_count+=1
    grad = gradient(x)
    grad_mag = np.linalg.norm(grad)
    x -= lr*grad
    x_hist_gd1 = np.hstack((x_hist_gd1, x.reshape(2,1).copy()))
    y_hist_gd1 = np.hstack((y_hist_gd1, loss_fcn(x)))

print('Total iterations: ', iter_count)
plt.figure(figsize=(10,5))
plt.plot(range(1,len(y_hist_gd1)+1), y_hist_gd1, '-o')
plt.title('Error History')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xlim([0,100])
plt.savefig("grad_descent_loss_lr={}.png".format(lr))

plot_loss_2d()
plt.plot(x_hist_gd1[0,:],x_hist_gd1[1,:],'m-o')
plt.savefig("grad_descent_2d_lr={}.png".format(lr))

ax=plot_loss_3d()
ax.plot(xs=x_hist_gd1[0,:],ys=x_hist_gd1[1,:],zs=y_hist_gd1, c='m')
plt.savefig("grad_descent_3d_lr={}.png".format(lr))
```

```python
# Gradient Descent - smaller learning rate
lr = 0.4
max_iter = 100
iter_count = 0
x = x_start.copy()
x_hist_gd2 = x.reshape(2,1).copy()
y_hist_gd2 = np.array([y_start])
grad_mag = float('inf')
while iter_count<max_iter and grad_mag>grad_thres:
    iter_count+=1
    grad = gradient(x)
    grad_mag = np.linalg.norm(grad)
    x -= lr*grad
    x_hist_gd2 = np.hstack((x_hist_gd2, x.reshape(2,1).copy()))
    y_hist_gd2 = np.hstack((y_hist_gd2, loss_fcn(x)))

print('Total iterations: ', iter_count)
plt.figure(figsize=(10,5))
plt.plot(range(1,len(y_hist_gd2)+1), y_hist_gd2, '-o')
plt.title('Error History')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xlim([0,100])
plt.savefig("grad_descent_loss_lr={}.png".format(lr))

plot_loss_2d()
plt.plot(x_hist_gd2[0,:],x_hist_gd2[1,:],'m-o')
plt.savefig("grad_descent_2d_lr={}.png".format(lr))

ax=plot_loss_3d()
ax.plot(xs=x_hist_gd2[0,:],ys=x_hist_gd2[1,:],zs=y_hist_gd2, c='m')
plt.savefig("grad_descent_3d_lr={}.png".format(lr))
```

```python
# Gradient Descent with Momentum
beta = 0.5
mu = 1.2
max_iter = 100
iter_count = 0
x = x_start.copy()
x_hist_moment = x.reshape(2,1).copy()
y_hist_moment = np.array([y_start])
grad_mag = float('inf')
v = np.array([0,0]).astype(float)
while iter_count<max_iter and grad_mag>grad_thres:
    iter_count+=1
    grad = gradient(x)
    grad_mag = np.linalg.norm(grad)
    # optimization - update velocity with gradient, then updadte weights with velocity
    v = beta*v + (1-beta)*grad
    x -= mu*v
    x_hist_moment = np.hstack((x_hist_moment, x.reshape(2,1).copy()))
    y_hist_moment = np.hstack((y_hist_moment, loss_fcn(x)))
print('Total iterations: ', iter_count)
plt.figure(figsize=(10,5))
plt.plot(range(1,len(y_hist_moment)+1), y_hist_moment, '-o')
plt.title('Error History')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xlim([0,100])
plt.savefig("momentum_loss.png")

plot_loss_2d()
plt.plot(x_hist_moment[0,:],x_hist_moment[1,:],'m-o')
plt.savefig("momentum_2d.png")

ax=plot_loss_3d()
ax.plot(xs=x_hist_moment[0,:],ys=x_hist_moment[1,:],zs=y_hist_moment, c='m')
plt.savefig("momentum_3d.png")
```

```python
# Gradient Descent with Momentum + Gyroscopic Force
beta = 0.5
mu = 1.2
# Gain for gyroscopic force
k = 0.1
max_iter = 100
iter_count = 0
x = x_start.copy()
x_hist_gyro = x.reshape(2,1).copy()
y_hist_gyro = np.array([y_start])
grad_mag = float('inf')
v = np.array([0,0]).astype(float)
while iter_count<max_iter and grad_mag>grad_thres:
    iter_count+=1
    grad = gradient(x)
    grad_mag = np.linalg.norm(grad)
    desired_v = x_goal-x
    curl = -k * np.cross(np.hstack([v, 0]), np.cross(np.hstack([desired_v, 0]), np.hstack([v, 0])))
    curl = curl[:-1]
    v = beta*v + (1-beta)*grad + curl
    x -= mu*v
    x_hist_gyro = np.hstack((x_hist_gyro, x.reshape(2,1).copy()))
    y_hist_gyro = np.hstack((y_hist_gyro, loss_fcn(x)))
print('Total iterations: ', iter_count)
plt.figure(figsize=(10,5))
plt.plot(range(1,len(y_hist_gyro)+1), y_hist_gyro, '-o')
plt.title('Error History')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xlim([0,100])
plt.savefig("gyroscopic_loss.png")

plot_loss_2d()
plt.plot(x_hist_gyro[0,:],x_hist_gyro[1,:],'m-o')
plt.savefig("gyroscopic_2d.png")

ax=plot_loss_3d()
ax.plot(xs=x_hist_gyro[0,:],ys=x_hist_gyro[1,:],zs=y_hist_gyro, c='m', linewidth=5)
# ax.view_init(70,315)
plt.savefig("gyroscopic_3d.png")
```
