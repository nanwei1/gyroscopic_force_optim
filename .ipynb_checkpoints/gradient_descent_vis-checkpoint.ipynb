```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# initialize the grid
A = np.array([[0.5,0],[0,2]])
x_start = np.array([-6,-6]).astype(float)
x_goal = np.array([0,0]).astype(float)
x1 = np.linspace(-7.5, 7.5, 150)
x2 = np.linspace(-7.5, 7.5, 150)
X1,X2 = np.meshgrid(x1, x2)
X = np.vstack((X1.reshape((1,-1)),X2.reshape((1,-1))))
X.shape
```

```python
# define loss function and compute loss for the grid
def loss_fcn(W, x):
    return x.T.dot(A.dot(x))/2
Y = np.array([])
for x in X.T:
    Y = np.append(Y, loss_fcn(A, x))
y_start = loss_fcn(A, x_start)
y_goal = loss_fcn(A, x_goal)
plt.scatter(X1,X2,c=Y.reshape(X1.shape))
```

```python
# 3d plot of the cost functions
from mpl_toolkits.mplot3d import axes3d
def plot_loss_3d():
    

    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    ax.contour3D(X1, X2, Y.reshape(X1.shape),100)
    start_plt = ax.scatter(x_start[0],x_start[1],y_start,s=50,c='r', marker = "o")
    goal_plt = ax.scatter(x_goal[0],x_goal[1],y_goal,s=50,c='r', marker = "x")
    ax.set_xlabel('$\mathregular{x_1}$')
    ax.set_ylabel('$\mathregular{x_2}$')
    ax.set_zlabel('y');
    ax.view_init(50, 20)
    ax.axis('equal')
    plt.legend([start_plt, goal_plt], ['start', 'goal'])
    ax.axis([-7,7,-7,7])
plt.savefig("loss_fcn_plot.png")
```

```python
# 2D contour plot of the cost function
# cp = plt.contour(X1,X2,Y.reshape(X1.shape))
# plt.clabel(cp, inline=1, fontsize=10)
# plt.axis('equal')
# plt.xlabel('$\mathregular{x_1}$')
# plt.ylabel('$\mathregular{x_2}$')
plt.figure(figsize=(10,10))
cp = plt.contour(X1,X2,Y.reshape(X1.shape), colors='k', linestyles='dashed')
plt.clabel(cp, inline=1, fontsize=10)
cp = plt.contourf(X1, X2, Y.reshape(X1.shape))
plt.xlabel('$\mathregular{x_1}$')
plt.ylabel('$\mathregular{x_2}$')
plt.axis('equal')
# plt.show()
```

```python
# define the gradient calculation
def gradient(W, x):
    return W.dot(x)
```

```python
lr = 0.8
max_iter = 100
iter_count = 0
x = x_start.copy()
x_hist = x.reshape(2,1).copy()
y_hist = np.array(x.T.dot(A.dot(x)))
grad_mag = float('inf')
grad_thres = 1e-2
while iter_count<max_iter and grad_mag>grad_thres:
    iter_count+=1
    grad = gradient(A,x)
    grad_mag = np.linalg.norm(grad)
    x -= lr*grad
    x_hist = np.hstack((x_hist, x.reshape(2,1).copy()))
    y_hist = np.hstack((y_hist, x.T.dot(A.dot(x))))
print('Total iterations: ', iter_count)
plt.plot(range(1,len(y_hist)+1), y_hist, '-o')
plt.title('Error History')
```

```python
plt.figure(figsize=(10,10))
cp = plt.contour(X1,X2,Y.reshape(X1.shape), colors='k', linestyles='dashed')
plt.clabel(cp, inline=1, fontsize=10)
cp = plt.contourf(X1, X2, Y.reshape(X1.shape))
start_plt = plt.scatter(x_start[0],x_start[1],s=100,c='r', marker = "^")
goal_plt = plt.scatter(x_goal[0],x_goal[1],s=100,c='r', marker = "x")
plt.plot(x_hist[0,:],x_hist[1,:],'m-o')
plt.xlabel('X1')
plt.ylabel('X2')
plt.axis('equal')
plt.show()
```

```python
beta = 0.6
mu = 0.2
max_iter = 100
iter_count = 0
x = x_start.copy()
x_hist = x.reshape(2,1).copy()
y_hist = np.array(x.T.dot(A.dot(x)))
grad_mag = float('inf')
grad_thres = 1e-2
v = np.array([0,0]).astype(float)
while iter_count<max_iter and grad_mag>grad_thres:
    iter_count+=1
    grad = gradient(A,x)
    grad_mag = np.linalg.norm(grad)
    v = beta*v + (1-beta)*grad
    x -= mu*v
    x_hist = np.hstack((x_hist, x.reshape(2,1).copy()))
    y_hist = np.hstack((y_hist, loss_fcn(A,x)))
print('Total iterations: ', iter_count)
print(y_hist)
```

```python
beta = 0.6
mu = 0.2
k = 0.01
max_iter = 100
iter_count = 0
x = x_start.copy()
x_hist = x.reshape(2,1).copy()
y_hist = np.array(x.T.dot(A.dot(x)))
grad_mag = float('inf')
grad_thres = 1e-2
v = np.array([0,0]).astype(float)
while iter_count<max_iter and grad_mag>grad_thres:
    iter_count+=1
    grad = gradient(A,x)
    grad_mag = np.linalg.norm(grad)
    desired_v = -x
    curl = -k * np.cross(np.hstack([v, 0]), np.cross(np.hstack([desired_v, 0]), np.hstack([v, 0])))
    curl = curl[:-1]
    v = beta*v + (1-beta)*grad + curl
    x -= mu*v
    x_hist = np.hstack((x_hist, x.reshape(2,1).copy()))
    y_hist = np.hstack((y_hist, loss_fcn(A,x)))
print('Total iterations: ', iter_count)
```

```python
plt.plot(range(1,len(y_hist)+1), y_hist, '-o')
plt.title("Error History")
```
